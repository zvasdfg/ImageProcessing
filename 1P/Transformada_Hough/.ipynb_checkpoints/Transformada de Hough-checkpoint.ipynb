# Transformada de Houg

## Ecuación de la recta

$Ax+By+C = 0$

$10x+4y+6=0$

$5x+2y+3=0$


$\hat{l} = (A,B,C)$

$\bar{\mathbf{x}}=(x,y,1)$

$\bar{\mathbf{x}} \cdot \hat{l} = Ax+By+C = 0$

$l = (\hat{n}_x,\hat{n}_y,d) = (\mathbf{\hat{n},d}), con ||\mathbf{\hat{n}}|| =1$

## Ejemplo

$3x+4y+5=0$

<img src = 'Recta.png' width = 600/>

$\hat{l} = (3,4,5)$

$|\hat{n}| = \sqrt{3^2+4^2} =\sqrt{9+16}=\sqrt{25}=5$

$l = (\frac{3}{5},\frac{4}{5},1)$

$(\hat{n}_x,\hat{n}_y) = (cos\theta, sin\theta)$

$cos\theta = 3/5$ entonces $\theta = arccos(3/5)\approx = 53.13º $


<img src = 'perpendicular.png' width = 600/>

$r = xcos(\theta) + y sin(\theta)$

Para cada punto $(x_0,y_0)$ existen una familia infinita de rectas que pasan por él, estas rectas están parametrizadas por $(r,\theta)$

Pensemos en el punto $(4,5)$, la familia de rectas que pasan por ese punto es: $r = 4cos(\theta)+5sin(\theta)$

## Restringiendo al espacio donde $r>0$ y $0<\theta<2\pi$


## Podemos trazar todas las curvas para todos los puntos de la imágen

Si dos o más curvas se cruza, significa que los puntos pertenecen a la misma recta

Se construye un acumulador para todos los cruces. Se da un threshold y aquellos que pasen el acumulador se considera que son rectas

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('sudoku.png')
```

```python
plt.figure(figsize=(15,15))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize = 3)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(edges,'gray')
```

```python
lines = cv2.HoughLinesP(edges,1,np.pi/180,10)

for line in lines:
    x1,y1,x2,y2 = line[0]
    cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

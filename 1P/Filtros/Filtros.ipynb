### Gaussian Blur

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
cat = cv2.imread('cat.jpeg')
```

```python
kernel = np.array([[1,4,6,4,1],[4,16,24,16,4],[6,24,36,24,6],[4,16,24,16,4],[1,4,6,4,1]])
kernel=kernel/256
kernel
```

```python
img_res = cv2.filter2D(src = cat,ddepth = -1,kernel = kernel)
img_res = cv2.filter2D(src = img_res,ddepth = -1,kernel = kernel)
img_res = cv2.filter2D(src = img_res,ddepth = -1,kernel = kernel)
img_res = cv2.filter2D(src = img_res,ddepth = -1,kernel = kernel)
img_res = cv2.filter2D(src = img_res,ddepth = -1,kernel = kernel)
img_res = cv2.filter2D(src = img_res,ddepth = -1,kernel = kernel)
img_res = cv2.filter2D(src = img_res,ddepth = -1,kernel = kernel)
img_res = cv2.filter2D(src = img_res,ddepth = -1,kernel = kernel)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(img_res,cv2.COLOR_BGR2RGB))
```

```python
img_res = cv2.GaussianBlur(cat,(201,201),cv2.BORDER_DEFAULT)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(img_res,cv2.COLOR_BGR2RGB))
```

### Median blur

```python
img = cv2.imread('salt_paper.png')
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
median = cv2.medianBlur(img,5)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(median,cv2.COLOR_BGR2RGB))
```

## Laplacian

## $\nabla^2 = \frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2} $

```python
np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
```

```python
img = cv2.imread('cat.jpeg')


laplacian = cv2.Laplacian(median,cv2.CV_64F)
plt.figure(figsize=(10,10))
plt.imshow(laplacian)
```

## Sobel

<img src = 'sobel_kernel.png' />

```python
img = cv2.imread('rubik2.jpeg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python

sobelx = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=5)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(sobelx,cmap='gray')
```

```python
plt.figure(figsize=(10,10))
plt.imshow(sobely,cmap='gray')
```

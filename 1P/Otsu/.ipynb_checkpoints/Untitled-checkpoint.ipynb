```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
im = cv2.imread("inputC.jpg",0)
[hist, _] = np.histogram(im, bins=256, range=(0, 255))
plt.hist(hist, bins=256, range=(0, 255))
print(im.shape[0])
print(im.shape[1])
print(im.shape[0]*im.shape[1])
```

```python
# Normalization so we have probabilities-like values (sum=1)
hist = hist/np.sum(hist)
#plt.hist(hist, bins=256, range=(0, 1))
val_max = 0
thr = 0
for t in range(1,255):
    q1 = np.sum(hist[:t])
    q2 = np.sum(hist[t:])
    m1 = np.sum(np.array([i for i in range(t)])*hist[:t])/q1
    m2 = np.sum(np.array([i for i in range(t,256)])*hist[t:])/q2
    val = q1*(1-q1)*np.power(m1-m2,2)
    if val_max < val:
        print('________________________')
        print(val_max)
        print(val)
        print(t)
        print(thr)
        print('________________________') 
        val_max = val
        thr = t
```

```python
plt.imshow(im, cmap = 'gray')
```

```python
print("Threshold: {}".format(thr))
```

```python
plt.imshow(im > thr, cmap = 'gray')
```

```python
ret2,thresh2 = cv2.threshold(im,127,255,cv2.THRESH_OTSU)
```

```python
plt.imshow(thresh2, cmap = 'gray')
print(ret2)
```

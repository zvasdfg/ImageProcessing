# Harris

### Weigthed sum squared difference

$\sum_i w(x_i)[I(x_i+u)-I(x_i)]^2$

### Auto correlation function

$\sum_i w(x_i)[I_0(x_i+\Delta u)-I_0(x_i)]^2$

### Recordando series de Taylor:

$f(x) \approx f(a)+f'(a)(x-a)$

$I(x_i+\Delta u) \approx I(x_i)+\nabla I(x_i)\cdot \Delta u$

$\sum_i w(x_i)[I(x_i+\Delta u)-I(x_i)]^2 \approx \sum_i w(x_i)[I(x_i)+\nabla I(x_i)\cdot \Delta u - I(x_i)]^2$

$=\sum_i w(x_i)[\nabla I(x_i)\cdot \Delta u ]^2$

$\sum_i w(x_i)[I_xu_x+I_yu_y]^2 = \sum_i w(x_i)[I_x^2u_x^2+2I_xI_yu_xu_y+I_y^2u_y^2]$

$A = \begin{bmatrix}
I_x^2 & I_xI_y\\
I_xI_y & I_y^2
\end{bmatrix}$

$= \begin{bmatrix}
u_x & u_y
\end{bmatrix}  \begin{bmatrix}
I_x^2 & I_xI_y\\
I_xI_y & I_y^2
\end{bmatrix}   \begin{bmatrix}
u_x \\ u_y
\end{bmatrix}$


$R = det(A)-ktrace(A)^2$

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('chess.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
res = cv2.cornerHarris(gray,2,3,0.04)
res = cv2.dilate(res,None)
img[res>0.005*res.max()]=[255,0,0]
```

```python
plt.imshow(res,'gray')
```

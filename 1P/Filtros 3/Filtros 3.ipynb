# Binarizaci√≥n

## Por threshold

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
sudoku = cv2.imread('sudoku.png')
sudoku = cv2.cvtColor(sudoku,cv2.COLOR_BGR2GRAY)
plt.imshow(sudoku,'gray')
```

```python
_, bw = cv2.threshold(sudoku, 120, 255, cv2.THRESH_BINARY )
plt.figure(figsize=(10,10))
plt.imshow(bw,'gray')
```

### Otsu

```python
_, bw = cv2.threshold(sudoku, 120, 255, cv2.THRESH_OTSU)
plt.figure(figsize=(10,10))
plt.imshow(bw,'gray')
```

<img src = 'otsu.png' width = 500/>


### Adaptativo

```python
bw = cv2.adaptiveThreshold(sudoku,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
plt.figure(figsize=(10,10))
plt.imshow(bw,'gray')
```

# Transformaciones geom√©tricas

```python
lena = cv2.imread('lena.jpg')
translation = np.float32([[1,0,-100],[0,1,100]])
res = cv2.warpAffine(lena,translation,lena.shape[0:2])

plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(res,cv2.COLOR_BGR2RGB))
```
